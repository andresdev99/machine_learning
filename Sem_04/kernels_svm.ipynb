## Kernel interpretation

```
import numpy as np
import matplotlib.pyplot as plt
# from mpl_toolkits import mplot3d
from PIL import Image
from sklearn.svm import SVC
from sklearn.datasets import make_gaussian_quantiles
```

```
X, y = make_gaussian_quantiles(cov=2,
                                 n_samples = 200 , n_features = 2   ,
                                 n_classes = 2 , random_state = 1)

X[y ==0] = X[y ==0] * 0.5
_,ax = plt.subplots(figsize=(6,5))
ax.scatter(X[:, 0] , X[:, 1] , c=y ,  cmap='Paired')
```

```
Z = X[:, 0] ** 2 + X[:, 1] ** 2
# Z = X[:, 0] * 2 + X[:, 1] * 2 + 1
```

```
f, _ = plt.subplots(figsize=(6,5))
ax = f.add_subplot(projection='3d')
ax.view_init(10, 30)



ax.scatter3D( X[:, 0] , X[:, 1] , Z , c = y , cmap='Paired')
ax.set_zlim(0, 5)

h_grid, v_grid = np.meshgrid(np.linspace(X[:, 0].min(), X[:, 0].max()), np.linspace(X[:, 1].min(), X[:, 1].max()))
z_grid = np.ones(h_grid.shape) * 1.5
ax.plot_surface(h_grid, v_grid, z_grid, alpha=0.5)
plt.tight_layout()
```
